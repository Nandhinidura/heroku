```python
import pandas as pd
import dash
import plotly.graph_objects as go
fig = go.Figure()
from jupyter_dash import JupyterDash
import dash_core_components as dcc
import dash_html_components as html

df1 = pd.read_csv('/content/IPL.csv')
df1_sorted = df1.sort_values(by='winner')
fig = go.Figure()
fig.add_trace(
    go.Bar(
        x=df1_sorted['winner'],
        y=df1_sorted['player_of_match'],
        name="Best Winning Team"
    ))

app = dash.Dash()
app.layout = html.Div([
    dcc.Graph(figure=fig)
])

app = JupyterDash()
app.layout =html.Div([
                      dcc.Graph(figure=fig)
])
app.run_server(debug=True)
```
